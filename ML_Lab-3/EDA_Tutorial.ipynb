## Create Sample Dataset

In this section, we create a simple sample dataset that will be used to demonstrate different EDA techniques and visualization plots.

The dataset includes both **numerical variables** (such as age, salary, and experience) and a **categorical variable** (department). This combination allows us to apply different types of analysis and visualization methods throughout the notebook.

## Univariate Analysis

Univariate analysis focuses on analyzing **a single variable** at a time.  
The goal is to understand the **distribution, central tendency, and spread** of the data.

### 📊 Histogram

A histogram shows the **distribution** of a **numerical variable**.

It helps us understand how the values are spread and whether the data is **skewed or normally distributed** by displaying the **frequency** of values **within intervals (bins).**

````python
sns.histplot(
    df["Salary"],        # The numerical variable we want to analyze
    bins=20,             # Number of bins (intervals) used to group the data
    color="#66a3e6",   # Color of the bars
    edgecolor="skyblue"  # Color of the borders around each bar
)

# Add a title to the plot
plt.title("Salary Distribution")

# Label the x-axis
plt.xlabel("Salary")

# Label the y-axis
plt.ylabel("Frequency")

# Display the plot
plt.show()
````

### 📊 Histogram with Density Curve

In this example, we also add a **KDE (Kernel Density Estimation) curve** to provide a smooth representation of the data distribution.  

This helps us better understand the overall shape of the data, such as whether it is **skewed, symmetric, or approximately normally** distributed.

````python
sns.histplot(
    df["Salary"],        # Numerical variable to analyze
    kde=True,            # Add a density curve on top of the histogram
    bins=20,             # Number of bins (intervals) used to group the data
    color="#66a3e6",   # Color of the bars
    edgecolor="skyblue"  # Color of the borders around each bar
)

# Add a title
plt.title("Salary Distribution")

# Label the y-axis
plt.ylabel("Frequency")

# Display the plot
plt.show()
````

### 📊 Distribution Plot

Distplot was previously used to visualize **data distributions** and **probability density**, but it has been deprecated in newer versions of Seaborn.

The recommended alternative is **histplot()** or **kdeplot()**.

````python
sns.distplot(
    df["Salary"],        # Numerical variable we want to analyze
    bins=20,             # Number of bins (intervals) used to group the data
    kde=True,            # Displays the density curve
    color="skyblue",     # Color of the histogram bars
)

# Add a title
plt.title("Salary Distribution")

# Display the plot
plt.show()
````

### 📈 KDE Plot

A KDE (Kernel Density Estimate) plot shows the **probability density** of a **numerical variable.**

It provides a **smoother view of the data distribution** compared to a histogram.

````python
sns.kdeplot(
    df["Salary"],          # Numerical variable we want to visualize
    fill=True,             # Fill the area under the curve
    color="#609CE5"      # Color of the density curve
)

# Add a title
plt.title("Salary Density Distribution")

# Display the plot
plt.show()
````

### 📍 Rug Plot

A rug plot displays **individual data points** along an axis. **Each small line** represents **one observation** in the dataset.

It is useful for seeing where values are concentrated.

````python
sns.rugplot(
    df["Salary"],        # Numerical variable being visualized
    color="darkgreen"    # Color of the rug marks
)

# Add a title
plt.title("Rug Plot of Salary")

# Display the plot
plt.show()
````

### 📈 KDE Plot with Rug Plot

This visualization combines a KDE plot to **estimate the distribution of the data** and a rug plot to **show the exact positions of individual data points.**

````python
sns.kdeplot(
    df["Salary"],          # Numerical variable we want to visualize
    fill=True,             # Fill the area under the curve
    color='#66a3e6'      # Color of the density curve
)
sns.rugplot(
    df["Salary"],        # Numerical variable being visualized
    color="darkgreen"    # Color of the rug marks
)

# Display the plot
plt.show()
````

### 📦 Box Plot

A box plot summarizes the distribution of a **numerical variable** using **quartiles.**

It helps detect **outliers**, understand the **spread** of the data, and identify the **median.**

````python
sns.boxplot(
    df["Salary"],      # Numerical variable displayed on the axis
    color="#7cb6f4"  # Color of the box
)

# Add a title
plt.title("Box Plot of Salary")

# Display the plot
plt.show()
````

### 🎻 Violin Plot

A violin plot combines a **box plot** and a **density plot.**

Help us to see the **spread**, **distribution shape**, and **concentration** of values.

````python
sns.violinplot(
    df["Salary"],          # Numerical variable being analyzed
    color="lightseagreen"  # Color of the violin shape
)

# Add a title
plt.title("Violin Plot of Salary")

# Display the plot
plt.show()
````

### 🏷️ Count Plot

A count plot is used to visualize the **frequency of categories** in a **categorical variable.**

It counts **how many times each category appears in the dataset** and displays the result as bars.

````python
sns.countplot(
    x="Department",      # Categorical variable to count
    data=df,             # Dataset used for the plot
    palette="pastel",    # Color palette used for the bars
    hue="Department"     # Color bars by category
)

# Add a title
plt.title("Department Distribution")

# Label the y-axis
plt.ylabel("Count")

# Display the plot
plt.show()
````

### 🥧 Pie Chart

A pie chart is used to show the **proportion** of **categories** in a dataset.

It helps us understand how each category contributes to the whole.

````python
# Count the number of observations in each category
department_counts = df["Department"].value_counts()

# Create the pie chart
plt.pie(
    department_counts,                               # Values representing the size of each slice
    labels=department_counts.index,                  # Category names displayed on the chart
    autopct="%1.1f%%",                               # Display percentage values on each slice
    colors=["#79aef3", "#ee814f", "#a8f3a7"], # Colors of the pie slices
    startangle=90                                    # Rotate the chart for better appearance
)

# Add a title
plt.title("Department Distribution")

# Display the plot
plt.show()
````

## Bivariate Analysis

Bivariate analysis examines the **relationship between two variables**.  
It helps identify **patterns, correlations, or trends** between them.

### 📊 Bar Plot

A bar plot **compares numerical values** across **different categories.**

**By default, Seaborn calculates the mean (average)** of the numerical variable for each category and displays it as bars.

````python
sns.barplot(
    x="Department",      # Categorical variable on the x-axis
    y="Salary",          # Numerical variable whose average will be calculated
    data=df,             # Dataset used for the plot
    palette="pastel",    # Color palette used for the bars
    hue="Department"     # Color bars by category
)

# Add a title
plt.title("Average Salary by Department")

# Label the y-axis
plt.ylabel("Average Salary")

# Display the plot
plt.show()
````

### 📦 Box Plot by Category

A box plot by category **compares the distribution** of a **numerical variable** across **different categories.**

It helps us identify differences in **medians**, **spread**, and **outliers** between groups.

````python
sns.boxplot(
    x="Department",      # Categorical variable on the x-axis
    y="Salary",          # Numerical variable on the y-axis
    data=df,             # Dataset used
    palette="Set3",      # Color palette
    hue="Department"     # Color by category
)

# Add a title
plt.title("Salary Distribution by Department")

# Display the plot
plt.show()
````

### 🎻 Violin Plot (by Category)

A violin plot **compares the distribution** of a **numerical variable** across **different categories.**

It combines features of a box plot and a density plot, allowing us to see the **spread**, **distribution shape**, and **concentration** of values within each category.

````python
sns.violinplot(
    x="Department",      # Categorical variable displayed on the x-axis
    y="Salary",          # Numerical variable displayed on the y-axis
    data=df,             # Dataset used for the plot
    palette="Set2",      # Color palette used for the violin shapes
    hue="Department"     # Color by category
)

# Add a title
plt.title("Salary Distribution by Department")

# Display the plot
plt.show()
````

### 🏷️ Count Plot with Hue

A count plot with hue allows us to compare the **frequency of categories** while **considering another categorical variable.**

This helps us understand how the distribution of categories changes across different groups.

````python
sns.countplot(
    x="Department",      # Main categorical variable
    data=df,             # Dataset used
    palette="flare",     # Color palette
    hue="Experience"     # Additional variable used to split categories
)

# Add a title
plt.title("Department Distribution by Experience")

# Label the y-axis
plt.ylabel("Count")

# Display the plot
plt.show()
````

### 🔹 Scatter Plot

A scatter plot shows the **relationship** between **two numerical variables.**

It helps us understand whether the relationship between the variables is **positive**, **negative**, or **nonexistent**.

````python
plt.scatter(
    df["Experience"],  # Variable on the x-axis
    df["Salary"],      # Variable on the y-axis
    color="teal",      # Color of the points
    alpha=0.6          # Transparency of the points
)

# Add a title
plt.title("Experience vs Salary")

# Label the axes
plt.xlabel("Experience")
plt.ylabel("Salary")

# Display the plot
plt.show()
````

### 🔹 Strip Plot

A strip plot shows the **distribution** of a **numerical variable** across **categories.**

Each data point is displayed as an individual dot, which helps **visualize how values are spread within each category.**

````python
sns.stripplot(
    x="Department",      # Categorical variable displayed on the x-axis
    y="Salary",          # Numerical variable displayed on the y-axis
    data=df,             # Dataset used for the plot
    palette="Set2",      # Color palette used for the categories
    hue="Department"     # Color by category
)

# Add a title
plt.title("Salary Distribution by Department")

# Display the plot
plt.show()
````

### 🐝 Swarm Plot

A swarm plot is similar to a strip plot but **arranges points so that they do not overlap.**

This makes it easier to see the exact distribution of values within each category.

````python
sns.swarmplot(
    x="Department",      # Categorical variable on the x-axis
    y="Salary",          # Numerical variable on the y-axis
    data=df,             # Dataset used
    palette="Set2",      # Color palette for the points
    hue="Department"     # Color by category
)

# Add a title
plt.title("Salary Distribution by Department")

# Display the plot
plt.show()
````

### 🔷 Hexbin Plot

A hexbin plot is used to visualize the **relationship** between **two numerical variables** when the **dataset is large.**

````python
plt.hexbin(
    df["Experience"],    # Variable on the x-axis
    df["Salary"],        # Variable on the y-axis
    gridsize=10,         # Number of hexagons used to divide the plot
    cmap="Blues"         # Color map representing density
)

# Add a color bar to show density scale
plt.colorbar(label="Count")

# Add a title
plt.title("Experience vs Salary")

# Label the axes
plt.xlabel("Experience")
plt.ylabel("Salary")

# Display the plot
plt.show()
````

### 📈 Regression Plot

An regression plot visualizes the **relationship** between **two numerical variables** and **fits a linear regression line** to show the overall trend in the data.

It helps us identify the **direction and strength of the relationship** between the variables.

````python
sns.regplot(
    x="Experience",          # Numerical variable on the x-axis
    y="Salary",              # Numerical variable on the y-axis
    data=df,                 # Dataset used to create the plot
    scatter_kws={"color":"steelblue"},  # Color of the scatter points
    line_kws={"color":"red"}            # Color of the regression line
)

# Add a title
plt.title("Regression Plot: Experience vs Salary")

# Display the plot
plt.show()
````

### 📈 Linear Model Plot (LM Plot)

A linear model plot provides the same regression visualization but **allows grouping the data by categories**, making it possible to compare regression trends across different groups.

````python
sns.lmplot(
    x="Experience",          # Numerical variable displayed on the x-axis
    y="Salary",              # Numerical variable displayed on the y-axis
    data=df,                 # Dataset used to create the plot
    hue="Department",        # Separate regression lines for each category
    scatter_kws={"color":"steelblue"},  # Color of scatter points
    col='Department'                    # Create a separate plot for each department
)

# Display the plot
plt.show()
````

### 🔗 Joint Plot

A joint plot visualizes the **relationship** between **two numerical variables.**

It **combines a plot such as (scatter, reg, resid, kde, and hex) with distribution plots** for each variable, helping us understand both the relationship and individual distributions.



````python
sns.jointplot(
    x="Experience",      # Numerical variable on the x-axis
    y="Salary",          # Numerical variable on the y-axis
    data=df,             # Dataset used
    kind="scatter",      # Type of plot (scatter, kde, hex, etc.)
    color="purple"       # Color of the points
)

# Display the plot
plt.show()
````

### 📈 Line Plot

A line plot is used to **visualize trends or changes** in a variable over another variable.

It is commonly used for **time series data** or to show general trends.

````python
sns.lineplot(
    x="Experience",      # Variable on the x-axis
    y="Salary",          # Variable on the y-axis
    data=df,             # Dataset used
    marker="o",          # Adds markers to highlight each data point
    color="teal",        # Color of the line
    errorbar=None        # Removes the confidence interval shading
)

# Add a title
plt.title("Salary Trend by Experience")

# Display the plot
plt.show()
````

## Multivariate Analysis

Multivariate analysis explores the **relationships among multiple variables simultaneously**.  
It helps uncover **complex patterns and interactions** within the data.

### 🔗 Correlation Matrix

A correlation matrix measures the **strength of relationships** between **numerical variables.**

It produces a **table of correlation** coefficients ranging from **-1 to 1**, indicating **negative, no, or positive relationships.**

````python
# Compute the correlation matrix
corr_matrix = df.corr(numeric_only=True)   # Calculates correlation between all numerical variables

# Display the correlation matrix
corr_matrix
````

### 🔥 Correlation Heatmap

A heatmap visualizes the **correlation matrix using colors**, making it easier to identify strong or weak relationships between variables.

````python
sns.heatmap(
    df.corr(numeric_only=True),   # Compute correlation between numerical variables
    annot=True,                   # Display correlation values inside each cell
    cmap="coolwarm",              # Color palette showing positive and negative correlations
    linewidths= 0.5               # Width of the lines separating the cells
)

# Add a title
plt.title("Correlation Heatmap")

# Display the plot
plt.show()
````

### 🧬 Cluster Map

A cluster map is an extension of a heatmap that **groups similar rows and columns together** using **hierarchical clustering.**

It helps identify **patterns and similarities within the data.**

````python
# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Create a cluster map
sns.clustermap(
    corr_matrix,        # Correlation matrix used for clustering
    cmap="coolwarm",    # Color palette
    annot=True,         # Display correlation values
    figsize=(8,6)       # Size of the figure
)

# Display the plot
plt.show()
````

### 🔗 Pair Plot

A pair plot shows **relationships** between **multiple numerical variables.**

It combines **scatter plots** and **distribution plots** for each variable pair.

````python
sns.pairplot(
    df,                # Dataset used
    hue="Department",  # Color points by category
    palette="Set2"     # Color palette
)

plt.show()
````

## Categorical Plot (Catplot)

A catplot is a **high-level function** in Seaborn used to **visualize categorical data.**  
It can generate different types of categorical plots by specifying the **`kind`** parameter.

The following plot types can be used with `kind`:

| Plot Name | Kind |
|----|----|
| Strip Plot | **"strip"** |
| Swarm Plot | **"swarm"** |
| Box Plot | **"box"** |
| Violin Plot | **"violin"** |
| Boxen Plot | **"boxen"** |
| Point Plot | **"point"** |
| Bar Plot | **"bar"** |
| Count Plot | **"count"** |

````python
# Create a categorical plot
sns.catplot(
    x="Department",      # Categorical variable on the x-axis
    y="Salary",          # Numerical variable on the y-axis
    data=df,             # Dataset used for the plot
    kind="bar",          # Type of categorical plot to generate
    hue="Department",    # Color by category
    palette="Set2",      # Color palette used for the plot
)

# Add a title
plt.title("Salary Distribution by Department")

# Display the plot
plt.show()
````

## 📝 Summary of EDA Visualizations

### Matplotlib vs Seaborn

| Feature | Matplotlib | Seaborn |
|--------|-----------|--------|
| Level | Low-level | High-level |
| Ease of use | More code required | Simpler and faster |
| Customization | Very flexible | Limited but sufficient |
| Default style | Basic | More visually appealing |
| Best for | Full control and customization | Statistical visualization and EDA |

In practice, both libraries are often used together to create clear and effective visualizations.


> **Note:**  
> **Matplotlib** provides the basic structure for creating plots, while **Seaborn** builds on top of it to create more advanced and visually appealing statistical visualizations.

### Univariate Analysis

Univariate analysis focuses on understanding the distribution and characteristics of a **single variable.**

| Rug Plot | `sns.rugplot()` | Displays individual data points along an axis |

### Bivariate Analysis

Bivariate analysis examines the relationship between **two variables.**

| Plot | Function | Purpose |
|-----|-----|-----|
| Bar Plot | `sns.barplot()` | Compares numerical values across categories |
| Scatter Plot | `plt.scatter()` / `sns.scatterplot()` | Shows relationship between two numerical variables |
| Strip Plot | `sns.stripplot()` | Shows individual data points across categories |

| Hexbin Plot | `plt.hexbin()` | Shows density for large scatter datasets |

Multivariate analysis explores relationships between **multiple variables.**

| Plot | Function | Purpose |
