### Basic Line Plot

A basic line plot is the **default type** of plot in Matplotlib.  

````python
# Create a new figure with specific size
plt.figure(figsize=(8,5))   # width=8, height=5

# Plot simple data
plt.plot([1,2,3], [4,5,6])  # x values, y values

# Add title
plt.title("Simple Line Plot")

# Label axes
plt.xlabel("X-axis")  # X-axis
plt.ylabel("Y-axis")  # Y-axis

# Display the plot
plt.show()
````

### Figure & Axes

A **figure** is the overall canvas that holds the plot, while **axes** represent the actual area where the data is drawn. 

This structure gives **more control** over how plots are created and customized.

````python
# Create figure and axes object
fig, axes = plt.subplots()   # fig = whole canvas, ax = actual plot

# Plot data on the axes
axes.plot([1,2,3], [4,5,6])  # plotting inside ax (not plt)

# Set title using axes
axes.set_title("Using Axes")

# Set labels
axes.set_xlabel("X-axis")
axes.set_ylabel("Y-axis")

# Display plot
plt.show()
````

### Subplots

Subplots allow us to create **multiple plots** within the **same figure.**  

This is useful for **comparing different visualizations** side by side.

````python
# Create multiple plots (nrows, ncols)
fig, axes = plt.subplots(ncols=2,nrows=1, figsize=(12,5))

# First subplot
axes[0].plot([1,2,3], [4,5,6])   # first plot
axes[0].set_title("Plot 1")

# Second subplot
axes[1].plot([1,2,3], [6,5,4])   # second plot
axes[1].set_title("Plot 2")

# Adjust spacing automatically
plt.tight_layout()

# Display plots
plt.show()
````

````python
# Generate random data
x = np.random.randint(3000, 15000, 100)   # 100 random values
y = np.random.randint(0, 10, 100)         # 100 random values

# Create multiple plots (nrows, ncols)
fig, axes = plt.subplots(1,2, figsize=(12,5))  # 2 plots side by side

# Histogram (distribution of x)
axes[0].hist(
    x,                   # Random salary data
    bins=20,             # Number of bins
)

# Add title to first plot
axes[0].set_title("Distribution")

# Scatter plot (relationship)
axes[1].scatter(
    x,              # X-axis
    y,              # Y-axis
)

# Add title to second plot
axes[1].set_title("X vs Y")

# Adjust spacing between plots
plt.tight_layout()

# Display plots
plt.show()
````

### Custom Axes (Inset Plot)

This example shows how to manually create **multiple axes** inside a **single figure.**
  
It allows us to place smaller plots **(inset plots)** within a larger main plot for better visualization.

````python
# Create a new figure (blank canvas)
fig = plt.figure(figsize=(8,5))  

# Add main axes → [left, bottom, width, height]
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])   # Main plot

# Add smaller inset axes inside the figure
axes2 = fig.add_axes([0.17, 0.5, 0.3, 0.3])   # Small plot (top-left)

# Plot on main axes
axes1.plot(
    [1,2,3], [4,5,6],      # x,y
    color="blue"           # Line color
)

# Customize main plot
axes1.set_xlabel("X")         # X label
axes1.set_ylabel("Y")         # Y label
axes1.set_title("Main Plot")  # Title

# Plot on inset axes (smaller plot)
axes2.plot(
    [4,5,6], [3,2,1],       # x,y
    color="red"             # Line color
)

# Customize inset plot
axes2.set_xlabel("X")         # X label
axes2.set_ylabel("Y")         # Y label
axes2.set_title("Inset Plot") # Title

# Display the figure
plt.show()
````

### Plot Customization

Customization allows us to **modify** the appearance of plots, such as **colors**, **line styles**, **markers**, and **grid lines.**  

This helps make visualizations clearer and more readable.



<div style="display: flex; gap: 20px;">

<div>

**Line Styling:**

| Parameter | Options | Description |
|----------|--------|-------------|
| `color` | `"red"`, `"blue"`, `"green"` | Line color |
| `linewidth` / `lw` | `0.5`, `1`, `2`, `3` | Line thickness |
| `linestyle` / `ls` | `'-'`, `'--'`, `':'`, `'-.'` | Line pattern |

</div>

<div>

**Marker Styling:**

| Parameter | Options | Description |
|----------|--------|-------------|
| `marker` | `'o'`, `'s'`, `'+'`, `'*'` | Marker shape |
| `markersize` | `2`, `4`, `8` | Marker size |
| `markerfacecolor` | `"red"`, `"yellow"` | Fill color |
| `markeredgecolor` | `"black"`, `"green"` | Border color |
| `markeredgewidth` | `1`, `2`, `3` | Border thickness |

</div>

</div>

````python
# Create figure
plt.figure(figsize=(8,5))

# Plot line with customization
plt.plot(
    [1,2,3], [4,5,6],
    color="red",        # line color
    linestyle=":",      # dashed line
    marker="o"          # points on line   
)

# Add title
plt.title("Customized Plot")

# Labels
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add grid
plt.grid(True)

# Show plot
plt.show()
````

````python
# Creat Random Data
x = np.linspace(0, 8, 11) 

# Create figure and axes
fig, ax = plt.subplots(figsize=(12,6))  # Set figure size

# Line Width (thickness):
ax.plot(x, x+1, color="red", linewidth=0.25)   # Very thin line
ax.plot(x, x+2, color="red", linewidth=0.5)    # Thin line
ax.plot(x, x+3, color="red", linewidth=1)      # Normal line
ax.plot(x, x+4, color="red", linewidth=2)      # Thick line


# Line Styles:
# '-'  = solid
# '--' = dashed
# '-.' = dash-dot
# ':'  = dotted
ax.plot(x, x+5, color="green", lw=3, linestyle='-')    # Solid line
ax.plot(x, x+6, color="green", lw=3, linestyle='-.')   # Dash-dot line
ax.plot(x, x+7, color="green", lw=3, linestyle=':')    # Dotted line


# Custom Dash Pattern:
line, = ax.plot(x, x+8, color="black", lw=1.5)  # Create line object
line.set_dashes([5, 10, 15, 10])  # Custom pattern: line, space, line, space


# Marker Types:
# '+' = plus
# 'o' = circle
# 's' = square
# '1' = triangle down
ax.plot(x, x+9,  color="blue", lw=3, ls='-',  marker='+')   # Plus marker
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')   # Circle marker
ax.plot(x, x+11, color="blue", lw=3, ls='-',  marker='s')   # Square marker
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')   # Triangle marker


# Marker Size & Colors:
ax.plot(x, x+13, color="purple", lw=1, marker='o', markersize=2)   # Small markers
ax.plot(x, x+14, color="purple", lw=1, marker='o', markersize=4)   # Medium markers
ax.plot(x, x+15, color="purple", lw=1, marker='o', markersize=8, 
        markerfacecolor="red")                                     # Large marker with fill color

ax.plot(x, x+16, color="purple", lw=1, marker='s', markersize=8, 
        markerfacecolor="yellow",      # Fill color
        markeredgewidth=3,             # Border thickness
        markeredgecolor="green")       # Border color


# Display plot
plt.show()
````

### Legend
A legend is used to **identify different elements in a plot**, such as multiple lines or categories.  

It helps **explain what each part of the visualization represents.**


**Legend Location (loc):**

| loc value | Position |
|----------|----------|
| `loc=1` | Upper Right |
| `loc=2` | Upper Left |
| `loc=3` | Lower Left |
| `loc=4` | Lower Right |

The `loc` parameter is used to control the position of the legend inside the plot.

````python
# Create a new figure
plt.figure(figsize=(8,5))

# First line
plt.plot(
        [1,2,3], [4,5,6],        # x,y
        label="Simple Line Plot" # Label for legend
)

# Second line
plt.plot(
        [1,2,3], [6,5,4],        # x,y 
        label="Reference Line"   # Another label
)

# Add legend to the plot
plt.legend(loc=2)   # Shows labels on the plot

# Titles and labels
plt.title("Legend Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Display
plt.show()
````

### Axis Limits  
Limits are used to **control the visible range of the axes.**
  
They help focus on specific parts of the data and improve visualization clarity.

````python
# Generate random data
x = np.random.rand(100) * 10      # Random values between 0 and 10
y = np.random.rand(100) * 10000   # Random values between 0 and 10000

# Create a new figure
plt.figure(figsize=(8,5))

# Scatter plot
plt.scatter(
    x, y,                 # Random data points
    color="red"           # Point color
)

# Set x-axis limits (zoom into part of the data)
plt.xlim(2, 8)            # Show only values between 2 and 8

# Set y-axis limits
plt.ylim(2000, 8000)      # Focus on mid-range values

# Add title and labels
plt.title("Axis Limits with Random Data")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Display the plot
plt.show()
````

### Figure Size & DPI

Matplotlib allows us to control the size and resolution of a figure.

- **figsize** → controls the width and height of the figure (in inches)  
- **dpi** → controls the resolution (dots per inch)

Adjusting these parameters helps improve the clarity and appearance of plots.

````python
# Create a new figure with custom size and resolution
fig = plt.figure(
    figsize=(8,4),   # Width = 8 inches, Height = 4 inches
    dpi=200
)

# Plot simple data
plt.plot(
    [1,2,3], [4,5,6],   # X and Y values
    color="blue",        # Line color
    marker='o'
)

# Add title and labels
plt.title("Figure Size & DPI Example")
plt.xlabel("X")
plt.ylabel("Y")


# Display the plot
plt.show()
````

### Saving Figure

Saving allows us to **export plots** as image files such as **PNG** or **PDF.**  

This is useful for reports, presentations, and documentation.

````python
# Create figure
plt.figure(figsize=(8,5))

# Plot data
plt.plot([1,2,3], [4,5,6])

# Save the figure before showing
plt.savefig("plot.png")   # saves image in your folder

# Show plot
plt.show()
````

## 📝 Summary of Matplotlib Basics

### Most Used Matplotlib Functions

| Function | Description |
|--------|------------|
| `plt.figure()` | Creates a new figure (plot area) |
| `plt.subplots()` | Creates figure and axes (used for multiple plots) |
| `plt.plot()` | Creates a line plot |
| `plt.scatter()` | Creates a scatter plot |
| `plt.hist()` | Creates a histogram |
| `plt.boxplot()` | Creates a box plot |
| `plt.title()` | Adds a title to the plot |
| `plt.xlabel()` | Labels the x-axis |
| `plt.ylabel()` | Labels the y-axis |
| `plt.legend()` | Displays legend for multiple lines |
| `plt.grid()` | Adds grid lines |
| `plt.xlim()` | Sets x-axis limits |
| `plt.ylim()` | Sets y-axis limits |
| `plt.tight_layout()` | Adjusts spacing between plots |
| `plt.savefig()` | Saves the figure as an image |
| `plt.show()` | Displays the plot |

### Matplotlib: plt vs axes

| Feature | `plt` (Pyplot) | `axes` (Axes) |
|--------|---------------|------------|
| Usage | Simple plots | Advanced / multiple plots |
| Target | Current figure | Specific subplot |
| Title | `plt.title()` | `axes.set_title()` |
| X label | `plt.xlabel()` | `axes.set_xlabel()` |
| Y label | `plt.ylabel()` | `axes.set_ylabel()` |
| Best for | Quick visualization | Professional and scalable code |

**plt** is easier for quick and simple plots, while **axes** provides better control and is preferred when working with multiple plots or advanced customization.


### Figure Size vs DPI

| Feature | figsize | dpi |
|--------|--------|-----|
| Definition | Controls the size of the figure | Controls the resolution of the figure |
| Unit | Inches | Dots per inch (pixels per inch) |
| Affects | Width and height of the plot | Clarity and sharpness of the plot |
| When to use | When the plot looks too small or large | When the plot looks blurry or unclear |
| Example | `figsize=(8,4)` | `dpi=100` |
| Result | Bigger or smaller plot | Clearer or more detailed plot |
